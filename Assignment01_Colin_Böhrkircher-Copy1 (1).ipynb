```python
# Assignment No.1 DigiBau3 
# Name: Colin
# Familyname: Böhrkircher


# Dimension of the Beam
D = 5                 # Dead load           (kN/m)
P = 7                 # Live load           (kN/m)
L = 3.2              # Length of the beam  (m)

# Elasticity Modul, yield strength, gamma coefficents for the following calculations
E = 210               # Elasticity Modul (GPa= kN/mm^2)
f_y= 235              # yield strength (MPa= N/mm²)
gamma_M0= 1.05        # gamma
gravity = 10          # gravitational acceleration (in m/s^2)
L_mm = L * 10**3      # Length of the beam in mm (mm)

# Informations stored in tuples for the calculations from the steel profile table givendddddd

IPE_Name  = ('IPE 80', 'IPE 120', 'IPE 160', 'IPE 200', 'IPE 270', 'IPE 300') #names of the IPE
IPE_Iy    = ( 801000 ,  3180000 , 8690000  , 19430000 , 57900000 , 83560000 ) #moment of inertia of major axis (mm^4)
IPE_G     = ( 6      , 10.4     , 15.8     , 22.4     , 36.1     , 42.2     ) # weight per length (kg/m)
IPE_A_vz  = ( 358    , 631      , 966      , 1400     , 2210     , 2570     ) # A_vz from the table (mm^2)
IPE_W_ply = ( 23200  , 60700    , 124000   , 221000   , 484000   , 628000   ) # W_pl.y from the table (mm^3)

# Define Function for the smallest verified IPE section for sls 

def Design_Beam(D,P,L):
    Q_sls = (D + P)/10**3              # Load Combination (Dead load + Live Load) (kN/mm)
    I_y   = (250*Q_sls*L_mm**3)/(8*E)  # I_y of the beam 
    return I_y

# Saving the Value for I_y from the steel beam

I_y = Design_Beam(D,P,L)       

# Saving the Data from the tuple IPE_IY in the variable

a=0                     #variable for position in the tuples 
I_IPE=IPE_Iy[a]                     

# Loop for the needed I_IPE of the beam
     
while I_y> I_IPE:
    a=a+1
    I_IPE=IPE_Iy[a]

# Result for the SLS of the beam

print('The verified steel section for the SLS is', IPE_Name[a])


# Calculating the G-force, max. shear force and max. bending moment for the beam with the IPE from above

G= IPE_G[a]*gravity/10**3     # G-force for the beam from the table (in N/mm)
Q_uls= (1.3*(D+G)+ 1.5*P)     # Load Combination for Q_ULS in (N/mm)
V_sup= Q_uls*L_mm             # shear force for the force Q_uls
M_sup= (Q_uls*L_mm**2)/2      # bending moment for the force Q_uls

# Calculating the shear force resistance and bending moment resistance  

W_ply= IPE_W_ply[a]                    # storing the value for W_ply for the beam
A_vz = IPE_A_vz[a]                      # storing the value for A_vz for the beam

V_cRD= (A_vz*(f_y/3**0.5))/ gamma_M0   # calculating the shear force resistance
M_cRD= (W_ply*f_y)/ gamma_M0           # calculating the bending moment resistance

while a< len(IPE_Name):                # loop for comparing the max. shear force with the shear force resistance 
    if V_sup<V_cRD:                    # and the max bending moment with the bending moment resistance for the beam 
        if M_sup<M_cRD:
            print('The verified steel section for the ULS is', IPE_Name[a])  #solution for the needed IPE-Beam for the ULS 
            break
    else :
        a=a+1                          # if the given IPE beam is has not enough resistance                 
                                       # we have to increase the IPE-Class by one step
```
